```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
```

## Load image 

```python
image = cv2.imread("zebra.jpg")
img_h, img_v, img_c = image.shape
image.shape
```

```python
plt.imshow(image)
```

```python
img_data = image.reshape([-1, 3])
img_data.shape
```

## Apply Gaussian Mixture Model

```python
gaussian_model = GaussianMixture(n_components=2, covariance_type="full", random_state=42)
gaussian_model.fit(img_data)
```

```python
labels = gaussian_model.predict(img_data)
labels.shape
```

```python
labels = labels.reshape([img_h, img_v, 1])
```

```python
plt.imshow(labels)
```

## Remove background

```python
labels = labels.reshape(image.shape[0], image.shape[1], 1)
segment = np.where(labels==0, 255, image).astype(np.uint8)
```

```python
plt.imshow(segment)
```
